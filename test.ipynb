```python
import yfinance as yf
```

```python
teststock = yf.Ticker("INFY.BO").history(period="max")
```

```python
teststock
```

```python
teststock.plot.line(y="Close", use_index = True)
```

```python
import xgboost as xgb
```

```python
train = teststock.iloc[:int(0.99*len(teststock)),:]
test = teststock.iloc[int(0.99*len(teststock)):,:]
features = ['Open','Volume']
target = ['Close']
```

```python
test
```

```python
model = xgb.XGBRegressor()
```

```python
model.fit(train[features],train[target])
```

```python
predictions = model.predict(test[features])
predictions
```

```python
accuracy = model.score(test[features], test[target])
accuracy
```
